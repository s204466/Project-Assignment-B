### Street Trees and Air Quality in NYC
The street trees in NYC not only provide aesthetic and social benefits, but also help to reduce the amount of carbon dioxide (CO2) in the air. As trees grow, they absorb CO2 through photosynthesis and store it in their trunks, branches, and leaves. According to research, a mature tree can absorb as much as 22 kg of CO2 per year [3]. With hundreds of thousands of street trees in NYC, these trees play a significant role in reducing the amount of CO2 in the air.

As we have seen above, the number of trees in New York City has increased by approximately 160,000 over the last two decades. This amounts to an average of 8,000 new trees each year. As a result, the trees have helped to remove around 37 million more kg of CO2 from the air than wouldn't have been possible without their presence. This significant reduction in CO2 levels highlights the essential role that street trees play in creating a healthier and more sustainable urban environment.

In addition to absorbing CO2, street trees also help to reduce air pollution. The leaves of the trees act as filters, trapping pollutants such as nitrogen oxides, sulfur dioxide, and particulate matter. Research has shown that street trees can reduce the amount of pollutants in the air by up to 60% [4].

The benefits of street trees are especially important in urban areas like NYC, where air pollution is a significant issue. According to the NYC Department of Health, air pollution contributes to over 3,000 premature deaths, more than 2,000 hospitalizations due to respiratory and cardiovascular causes, and approximately 6,000 emergency department visits for asthma each year in the city [5]. By planting and maintaining street trees, NYC is not only improving the aesthetic and social environment, but also helping to reduce the negative health impacts of air pollution.
