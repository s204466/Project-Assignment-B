```python
import numpy as np
import pandas as pd
import folium
from folium import plugins
from folium.plugins import HeatMap

data = pd.read_csv("../Police_Department_Incident_Reports__Historical_2003_to_May_2018.csv")

data.head()
```

```python
mapSF1 = folium.Map(location=[37.77, -122.40],
                    zoom_start = 13)
mapSF1
```

```python
mapSF2 = folium.Map(location=[37.77, -122.40],
                    tiles = "Stamen Toner",
                    zoom_start = 13)

folium.Marker([37.77919, -122.41914],
              popup='City Hall',
              icon=folium.Icon(color='blue')
             ).add_to(mapSF2)

mapSF2
```

```python
#Here I've chosen jan/feb 2015 as the time interval
months = ["01", "02"]

dataPopups = data[data["Date"].str.contains("2015") == True]
dataPopups = dataPopups[dataPopups["Date"].str[:-8].str.contains('|'.join(months)) == True]
dataPopups = dataPopups[dataPopups["Category"].str.contains("DRUG/NARCOTIC") == True]
dfPopup = pd.DataFrame(dataPopups)
```

```python
mapSF3 = folium.Map(location=[37.77, -122.40],
                        tiles = "Stamen Toner",
                    zoom_start = 13) 

for i in range(0, len(dfPopup)):
    folium.CircleMarker(
        location=[dfPopup.iloc[i]['Y'], dfPopup.iloc[i]['X']],
        radius=1,
        color='red',
    ).add_to(mapSF3)

mapSF3
```

```python
dataHeatmap = data[data["Category"].str.contains("SEX OFFENSES, NON FORCIBLE") == True]
dfHeatmap = pd.DataFrame(dataHeatmap)

mapSF4 = folium.Map(location=[37.77, -122.40],
                        tiles = "Stamen Toner",
                    zoom_start = 13) 

dfHeatmap['Y'] = dfHeatmap['Y'].astype(float)
dfHeatmap['X'] = dfHeatmap['X'].astype(float)

dfHeat = dfHeatmap[['Y', 'X']]
dfHeat = dfHeat.dropna(axis=0, subset=['Y','X'])

dataHeat = [[row['Y'],row['X']] for index, row in dfHeat.iterrows()]

HeatMap(dataHeat).add_to(mapSF4)

mapSF4
```

The difference between a scatter-plot and a heatmap is that with the scatterplot we can easily see the precise location for each datapoint, but if there are multiple datapoints at the exact same location, we cant see how many there are. <br />
With the heatmap it is more difficult to see the precise location for each datepoint, but if there are more than one datapoint at a specific location, the color of the heatmap in that area will be more intense, and thus we can see if that area is a hotspot.

```python
dataTime = data[data["Category"].str.contains("DRIVING UNDER THE INFLUENCE") == True]
dfTime = pd.DataFrame(dataTime)

mapSF5 = folium.Map(location=[37.77, -122.40],
                        tiles = "Stamen Toner",
                    zoom_start = 13) 

dfTime['Y'] = dfTime['Y'].astype(float)
dfTime['X'] = dfTime['X'].astype(float)

dfTime2 = dfTime[['Y', 'X']]

dfTime2['Time'] = dfTime['Time'].str[0:2]
dfTime2['Time'] = dfTime2['Time'].astype(float)
dfTime2 = dfTime2.dropna(axis=0, subset=['Y','X', 'Time'])

dfTime2 = [[[row['Y'],row['X']] for index, row in dfTime2[dfTime2['Time'] == i].iterrows()] for i in range(0,23)]

hm = plugins.HeatMapWithTime(dfTime2,auto_play=True,max_opacity=0.8)
hm.add_to(mapSF5)
mapSF5
```

It can be seen in the heatmap, at what hours of the day there are a lot of "DRIVING UNDER THE INFLUENCE" cases (and it can be seen that it is roughly between 18:00 - 03:00)

I chose the "DRIVING UNDER THE INFLUENCE" crimetype and categorized it by hours of the day because this has a very high contrast of datapoints throughout the day, so we can see a lot of change in the heatmap over time.
