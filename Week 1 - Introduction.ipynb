# Exercise 1

* Report the total number of crimes in the dataset.
* List the various categories of crime. How many are there?
* List the number of crimes in each category

First is to import the data from the .csv file 

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt

data = pd.read_csv("../Police_Department_Incident_Reports__Historical_2003_to_May_2018.csv")

data.head()
```

```python
# Count total number of crimes (computed as the total number of rows in the csv file) 

print("Total number of crimes: " + str(len(data.index)))
```

```python
# List and count the number of different crime categories
print("The number of crime categories: " + str(data["Category"].nunique()) + "\n")

print("The different crime categories: \n" + str(data["Category"].unique()))

```

```python
print("The number of crimes commited in each category: \n" + str(data.groupby(["Category"]).size()))
```

# Exercise 2
* We have already counted the number of crimes in each category. What is the most commonly occurring category of crime? What is the least frequently occurring?
* Create a bar-plot over crime occurrences. This is a data visualization class, so here is the first essential lesson: For a plot to be informative you need to label the axes (The police chief will be furious if you forget). It can also be nice to other relevant pieces of info, title, labels, etc.).


```python
print("The number of crimes commited in each category in descending order: \n" + str(data['Category'].value_counts().sort_values(ascending=False)))
```

```python
plotData = data['Category'].value_counts().sort_values(ascending=False)

plotData.plot.bar()
plt.xticks(rotation=90, horizontalalignment="center")
plt.title("Occurences of each crime type")
plt.xlabel("Category")
plt.ylabel("Occurrences")
```
